#### **Accuracy**

As the *n of correct predictions / total size* 

**(TP + TN) / (TP + TN + FP + FN)**

```python
wg['accuracy'] = wg.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
wg_cov['accuracy'] = wg_cov.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
peu['accuracy'] = peu.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
peu_cov['accuracy'] = peu_cov.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
ped['accuracy'] = ped.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
ped_cov['accuracy'] = ped_cov.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
b2['accuracy'] = b2.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
b2_cov['accuracy'] = b2_cov.apply(lambda row: (row.TP + row.TN)/(row.TP + row.TN + row.FP + row.FN), axis=1)
```

#### **Precision**

Measure of how many of the positive predictions made are correct (TP).

As the *n of correctly predicted positive instances / n of total positive predictions*

**TP / (TP + FP)**

```python
wg['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in wg[['TP', 'FP']].values]
wg_cov['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in wg_cov[['TP', 'FP']].values]
peu['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in peu[['TP', 'FP']].values]
peu_cov['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in peu_cov[['TP', 'FP']].values]
ped['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in ped[['TP', 'FP']].values]
ped_cov['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in ped_cov[['TP', 'FP']].values]
b2['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in b2[['TP', 'FP']].values]
b2_cov['precision'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in b2_cov[['TP', 'FP']].values]
```

#### **Recall / Sensitivity**

Measure of how many of the positive cases (TP) the classifier correctly predicted, over all the positive cases (TP + FN)

**TP/(TP+FN)**

```python
wg['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in wg[['TP', 'FN']].values]
wg_cov['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in wg_cov[['TP', 'FN']].values]
peu['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in peu[['TP', 'FN']].values]
peu_cov['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in peu_cov[['TP', 'FN']].values]
ped['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in ped[['TP', 'FN']].values]
ped_cov['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in ped_cov[['TP', 'FN']].values]
b2['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in b2[['TP', 'FN']].values]
b2_cov['recall'] = [0 if x == 0
                 else x/(x+y)
                 for x,y in b2_cov[['TP', 'FN']].values]
```

#### **F1-Score**

Measure combining the precision and recall - weights the two ratios in a balanced way.

**2 x ((Precision x Recall) / (Precision + Recall))**

```python
def f1(p, r):
    return 2*((p*r)/(p+r))
```

```python
wg['F1-score'] = f1(wg['precision'],wg['recall'])
wg_cov['F1-score'] = f1(wg_cov['precision'],wg_cov['recall'])
peu['F1-score'] = f1(peu['precision'],peu['recall'])
peu_cov['F1-score'] = f1(peu_cov['precision'],peu_cov['recall'])
ped['F1-score'] = f1(ped['precision'],ped['recall'])
ped_cov['F1-score'] = f1(ped_cov['precision'],ped_cov['recall'])
b2['F1-score'] = f1(b2['precision'],b2['recall'])
b2_cov['F1-score'] = f1(b2_cov['precision'],b2_cov['recall'])
```

```python
wg.sort_values('Size', ascending=False, inplace=True)
wg_cov.sort_values('Size', ascending=False, inplace=True)
peu.sort_values('Size', ascending=False, inplace=True)
peu_cov.sort_values('Size', ascending=False, inplace=True)
ped.sort_values('Size', ascending=False, inplace=True)
ped_cov.sort_values('Size', ascending=False, inplace=True)
b2.sort_values('Size', ascending=False, inplace=True)
b2_cov.sort_values('Size', ascending=False, inplace=True)
```
